Given a linked list ofÂ **N**Â nodes such that it may contain a loop.

A loop here means that the last node of the link list is connected to the node at position X(1-based index). If the link list does not have any loop, X=0.

Remove the loop from the linked list, if it is present, i.e. unlink the last node which is forming the loop.

**Example 1:**

Input:
N = 3
value[] = {1,3,4}
X = 2
Output:1
Explanation:The link list looks like
1 -> 3 -> 4
     ^    |
     |____|
A loop is present. If you remove it
successfully, the answer will be 1.

**Example 2:**

Input:
N = 4
value[] = {1,8,3,4}
X = 0
Output:1
Explanation:TheÂ Linked list does not
contains any loop.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def detectAndRemoveLoop(head):
    if not head or not head.next:
        return
    
    slow = head
    fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            break

    if slow != fast:
        return

    slow = head
    while slow.next != fast.next:
        slow = slow.next
        fast = fast.next

    fast.next = None

# Create a linked list with a loop: 1 -> 2 -> 3 -> 4 -> 5 -> 3
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = head.next.next

# Remove the loop
detectAndRemoveLoop(head)

# Traverse the linked list after removing the loop
curr = head
while curr:
    print(curr.data, end=" ")
    curr = curr.next

# Output should be: 1 2 3 4 5
```

ðŸ’¡ **Question 2**

A numberÂ **N**Â is represented in Linked List such that each digit corresponds to a node in linked list. You need to add 1 to it.

**Example 1:**

Input:
LinkedList: 4->5->6
Output:457

**Example 2:**
    
Input:
LinkedList: 1->2->3
Output:124

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def addOne(head):
    dummy = ListNode(0)  
    dummy.next = head
    curr = dummy
    carry = 1

    while curr.next:
        curr = curr.next
        curr.val += carry
        if curr.val < 10:
            carry = 0
        else:
            curr.val %= 10
            carry = 1

    if carry == 1:
        new_node = ListNode(1)
        new_node.next = dummy.next
        dummy.next = new_node

    return dummy.next

# Example usage:
# Input: 9 -> 9 -> 9
# Output: 1 -> 0 -> 0 -> 0
head = ListNode(9)
head.next = ListNode(9)
head.next.next = ListNode(9)

new_head = addOne(head)

curr = new_head
while curr:
    print(curr.val, end=" -> ")
    curr = curr.next
print("None")
```

ðŸ’¡ **Question 3**

Given a Linked List of size N, where every node represents a sub-linked-list and contains two pointers:(i) aÂ **next**Â pointer to the next node,(ii) aÂ **bottom**Â pointerÂ to a linked list where this node is head.Each of theÂ sub-linked-list is in sorted order.Flatten the Link List such that all the nodes appear in a single level while maintaining the sorted order.Â **Note:**Â The flattened list will be printed using the bottom pointer instead of next pointer.

**Example 1:**

Input:
5 -> 10 -> 19 -> 28
|     |     |     |
7     20    22   35
|           |     |
8          50    40
|                 |
30               45
Output:Â 5-> 7-> 8- > 10 -> 19-> 20->
22-> 28-> 30-> 35-> 40-> 45-> 50.
Explanation:
The resultant linked lists has every
node in a single level.(Note:| represents the bottom pointer.)

```python
class ListNode:
    def __init__(self, data=None, next=None, bottom=None):
        self.data = data
        self.next = next
        self.bottom = bottom

def mergeLists(list1, list2):
    # Base cases
    if not list1:
        return list2
    if not list2:
        return list1

    # Choose the smaller value as the head of the merged list
    if list1.data < list2.data:
        result = list1
        result.bottom = mergeLists(list1.bottom, list2)
    else:
        result = list2
        result.bottom = mergeLists(list1, list2.bottom)

    return result

def flattenLinkedList(head):
    if not head or not head.next:
        return head

    # Recursively flatten the remaining linked list
    head.next = flattenLinkedList(head.next)

    # Merge the current linked list with the flattened next linked list
    head = mergeLists(head, head.next)

    return head

# Example usage:
# Linked list with three sub-linked lists: 5->7->8->30, 10->20, 19->22->50
head = ListNode(5)
head.next = ListNode(10)
head.next.next = ListNode(19)

# Set up the bottom pointers for the sub-linked lists
head.bottom = ListNode(7)
head.bottom.bottom = ListNode(8)
head.bottom.bottom.bottom = ListNode(30)

head.next.bottom = ListNode(20)

head.next.next.bottom = ListNode(22)
head.next.next.bottom.bottom = ListNode(50)

flattened_head = flattenLinkedList(head)

# Traverse the flattened linked list and print the values
curr = flattened_head
while curr:
    print(curr.data, end=" -> ")
    curr = curr.bottom
print("None")
```

ðŸ’¡ **Question 4**

You are given a specialÂ linked listÂ withÂ **N**Â nodes where each node has aÂ next pointer pointing to itsÂ next node. You are also givenÂ **M**Â random pointers, where you will be givenÂ **M**Â number of pairs denoting two nodesÂ **a**Â andÂ **b**Â Â **i.e.Â a->arb = b**Â (arb is pointer to random node)**.**

Construct a copy of the given list. The copy should consist of exactlyÂ **N**Â new nodes, where each new node has its value set to the value of its corresponding original node. Both the next and random pointer of the new nodes should point to new nodes in the copied list such that the pointers in the original list and copied list represent the same list state. None of the pointers in the new list should point to nodes in the original list.

For example, if there are two nodesÂ **X**Â andÂ **Y**Â in the original list, whereÂ **X.arb**Â **-->**Â **Y**, then for the corresponding two nodesÂ **x**Â andÂ **y**Â in the copied list,Â **x.arb --> y.**

Return the head of the copied linked list.

**Example 1:**

Input:
N = 4, M = 2
value = {1,2,3,4}
pairs = {{1,2},{2,4}}
Output:1
Explanation:In this test case, there
are 4 nodes in linked list.Â  Among these
4 nodes,Â  2 nodes have arbitrary pointer
set, rest two nodes have arbitrary pointer
as NULL. Second line tells us the value
of four nodes. The third line gives the
information about arbitrary pointers.
The first node arbitrary pointer is set to
node 2.Â  The second nodeÂ arbitrary pointer
is set to node 4.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.random = None

def copyRandomList(head):
    if not head:
        return None

    # Step 1: Create a copy of each node and insert it next to the original node
    curr = head
    while curr:
        new_node = Node(curr.data)
        new_node.next = curr.next
        curr.next = new_node
        curr = new_node.next

    # Step 2: Set the random pointers of the copied nodes
    curr = head
    while curr:
        if curr.random:
            curr.next.random = curr.random.next
        curr = curr.next.next

    # Step 3: Separate the original and copied linked lists
    curr = head
    new_head = head.next
    while curr.next:
        temp = curr.next
        curr.next = temp.next
        curr = temp

    return new_head

# Example usage:
# Original linked list: 7 -> 4 -> 5 -> 2 -> None
head = Node(7)
head.next = Node(4)
head.next.next = Node(5)
head.next.next.next = Node(2)

# Set up random pointers
head.random = head.next.next
head.next.random = head.next.next.next
head.next.next.random = head.next

# Create a copy of the linked list
copied_head = copyRandomList(head)

# Traverse the copied linked list and print the values and random pointers
curr = copied_head
while curr:
    print("Value:", curr.data, end=" ")
    if curr.random:
        print("Random Pointer:", curr.random.data)
    else:
        print("Random Pointer: None")
    curr = curr.next
```

ðŸ’¡ **Question 5**

Given theÂ `head`Â of a singly linked list, group all the nodes with odd indices together followed by the nodes with even indices, and returnÂ *the reordered list*.

TheÂ **first**Â node is consideredÂ **odd**, and theÂ **second**Â node isÂ **even**, and so on.

Note that the relative order inside both the even and odd groups should remain as it was in the input.

You must solve the problemÂ inÂ `O(1)`Â extra space complexity andÂ `O(n)`Â time complexity.

**Example 1:**

Input: head = [1,2,3,4,5]
Output: [1,3,5,2,4]

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def oddEvenList(head):
    if not head or not head.next:
        return head

    odd_head = head
    even_head = head.next
    odd = odd_head
    even = even_head

    while even and even.next:
        odd.next = even.next
        odd = odd.next
        even.next = odd.next
        even = even.next

    odd.next = even_head

    return odd_head

# Create the linked list from the input values
head = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(3)
node4 = ListNode(4)
node5 = ListNode(5)

head.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

# Call the function with the created linked list
result_head = oddEvenList(head)

# Print the values of the modified linked list
current = result_head
while current:
    print(current.val, end=" ")
    current = current.next
```

ðŸ’¡ **Question 6**

Given a singly linked list of sizeÂ **N**. The task is toÂ **left-shift**Â the linked list byÂ **k**Â nodes, whereÂ **k**Â is a given positive integer smaller than or equal to length of the linked list.

**Example 1:**

Input:
N = 5
value[] = {2, 4, 7, 8, 9}
k = 3
Output:8 9 2 4 7
Explanation:Rotate 1:4 -> 7 -> 8 -> 9 -> 2
Rotate 2:Â 7Â -> 8Â -> 9Â -> 2Â -> 4
Rotate 3:Â 8Â -> 9Â -> 2Â -> 4Â -> 7

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def leftShiftLinkedList(head, k):
    if not head or k <= 0:
        return head

    # Find the length of the linked list
    length = 0
    current = head
    while current:
        length += 1
        current = current.next

    # Adjust k to be within the range of the list length
    k = k % length

    # If k is 0, no need to shift
    if k == 0:
        return head

    # Find the (k+1)th node from the beginning
    prev = None
    current = head
    for _ in range(k):
        prev = current
        current = current.next

    # Set the new head and tail of the shifted list
    new_head = current
    new_tail = current
    while new_tail.next:
        new_tail = new_tail.next

    # Connect the original tail to the original head
    new_tail.next = head

    # Set the next node of the (k+1)th node as the new tail
    prev.next = None

    return new_head

# Create the linked list from the input values
head = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(3)
node4 = ListNode(4)
node5 = ListNode(5)

head.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

# Shift the linked list by k nodes
k = 2
shifted_head = leftShiftLinkedList(head, k)

# Print the values of the shifted linked list
current = shifted_head
while current:
    print(current.val, end=" ")
    current = current.next
```

ðŸ’¡ **Question 7**

You are given theÂ `head`Â of a linked list withÂ `n`Â nodes.

For each node in the list, find the value of theÂ **next greater node**. That is, for each node, find the value of the first node that is next to it and has aÂ **strictly larger**Â value than it.

Return an integer arrayÂ `answer`Â whereÂ `answer[i]`Â is the value of the next greater node of theÂ `ith`Â node (**1-indexed**). If theÂ `ith`Â node does not have a next greater node, setÂ `answer[i] = 0`.

**Example 1:**

Input: head = [2,1,5]
Output: [5,5,0]

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def nextLargerNodes(head):
    # Convert the linked list to a list
    values = []
    current = head
    while current:
        values.append(current.val)
        current = current.next

    # Initialize the answer list with 0s
    answer = [0] * len(values)
    stack = []

    # Process each value in reverse order
    for i in range(len(values)-1, -1, -1):
        # Pop elements from the stack until a greater element is found
        while stack and stack[-1] <= values[i]:
            stack.pop()

        # If a greater element is found, assign it to the answer list
        if stack:
            answer[i] = stack[-1]

        # Push the current value to the stack
        stack.append(values[i])

    return answer


# Create the linked list from the input values
head = ListNode(2)
node2 = ListNode(7)
node3 = ListNode(4)
node4 = ListNode(3)
node5 = ListNode(5)

head.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

# Find the next greater nodes
result = nextLargerNodes(head)

# Print the values of the next greater nodes
for value in result:
    print(value, end=" ")
```

ðŸ’¡ **Question 8**

Given theÂ `head`Â of a linked list, we repeatedly delete consecutive sequences of nodes that sum toÂ `0`Â until there are no such sequences.

After doing so, return the head of the final linked list.Â  You may return any such answer.

(Note that in the examples below, all sequences are serializations ofÂ `ListNode`Â objects.)

**Example 1:**

Input: head = [1,2,-3,3,1]
Output: [3,1]
Note: The answer [1,2,1] would also be accepted.

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def removeZeroSumSublists(head):
    # Create a dummy node to handle edge cases
    dummy = ListNode(0)
    dummy.next = head
    prefix_sum = 0
    prefix_sums = {0: dummy}

    # Traverse the linked list and compute prefix sums
    current = dummy.next
    while current:
        prefix_sum += current.val

        # If the current prefix sum exists in the prefix sums dictionary,
        # it means there is a subarray between the last occurrence of that
        # prefix sum and the current node whose sum is 0.
        # Remove that subarray from the linked list.
        if prefix_sum in prefix_sums:
            prev = prefix_sums[prefix_sum]
            prev.next = current.next

            # Update the prefix sums dictionary by removing the nodes
            # between the last occurrence of the prefix sum and the current node.
            temp = prev.next
            while temp != current:
                prefix_sum += temp.val
                del prefix_sums[prefix_sum]
                temp = temp.next

            # Continue the loop with the current node removed
            current = prev.next
        else:
            # Add the current prefix sum and node to the prefix sums dictionary
            prefix_sums[prefix_sum] = current
            current = current.next

    return dummy.next


# Create the linked list from the input values
head = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(-3)
node4 = ListNode(3)
node5 = ListNode(1)

head.next = node2
node2.next = node3
node3.next = node4
node4.next = node5

# Remove consecutive sequences of nodes that sum to 0
result = removeZeroSumSublists(head)

# Print the final linked list
current = result
while current:
    print(current.val, end=" ")
    current = current.next
```
