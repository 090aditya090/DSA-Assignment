Given a binary tree, your task is to find subtree with maximum sum in tree.

Output1 : 28

As all the tree elements are positive, the largest subtree sum is equal to sum of all tree elements.

Input2 :

     /Â  Â  \

  -2Â  Â  Â  3

  / \Â  Â  /Â  \

4 Â  5Â  -6 Â  2

Output2 : 7

Subtree with largest sum is :

 -2

 / \

4 Â  5

Also, entire tree sum is also 7.


```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def findMaxSubtreeSum(root):
    max_sum = float('-inf')  # Initialize maximum sum as negative infinity
    max_subtree = None  

    def dfs(node):
        nonlocal max_sum, max_subtree

        if node is None:
            return 0

        left_sum = dfs(node.left)
        right_sum = dfs(node.right)
        subtree_sum = left_sum + right_sum + node.val

        if subtree_sum > max_sum:
            max_sum = subtree_sum
            max_subtree = node

        return subtree_sum

    dfs(root)
    return max_subtree


root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

max_subtree = findMaxSubtreeSum(root)
print("Maximum subtree sum:", max_subtree.val)
```

Construct the BST (Binary Search Tree) from its given level order traversal.

Example:

Input: arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output: BST:

            7

         /Â  Â  \

       4 Â  Â  12

     /Â  \ Â  Â  /

    3 Â  6Â  8

   / Â   / Â    \

 1Â    5  Â    10

</aside>

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

def insertNode(root, val):
    if root is None:
        return TreeNode(val)

    if val < root.val:
        root.left = insertNode(root.left, val)
    else:
        root.right = insertNode(root.right, val)

    return root

def constructBST(level_order):
    if not level_order:
        return None

    root = TreeNode(level_order[0])
    for i in range(1, len(level_order)):
        insertNode(root, level_order[i])

    return root


level_order = [7, 4, 12, 3, 6, 8, 1, 5, 10]

root = constructBST(level_order)


def inOrderTraversal(node):
    if node is not None:
        inOrderTraversal(node.left)
        print(node.val, end=' ')
        inOrderTraversal(node.right)

print("In-order traversal of constructed BST:")
inOrderTraversal(root)
```

Given an array of size n. The problem is to check whether the given array can represent the level order traversal of a Binary Search Tree or not.

Examples:

Input1 : arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output1 : Yes

For the given arr[], the Binary Search Tree is:

            7

         /Â  Â  \

       4 Â  Â  12

     /Â  \ Â  Â  /

    3 Â  6Â  8

   / Â   / Â    \

Input2 : arr[] = {11, 6, 13, 5, 12, 10}

Output2 : No

The given arr[] does not represent the level order traversal of a BST.

```python
def isLevelOrderBST(level_order):
    if not level_order:
        return True

    n = len(level_order)

    for i in range(int(n/2) - 1, -1, -1):
        left_child = 2*i + 1
        right_child = 2*i + 2

        if level_order[i] > level_order[left_child] or (right_child < n and level_order[i] > level_order[right_child]):
            return False

    return True

level_order = [7, 4, 12, 3, 6, 8, 1, 5, 10]

if isLevelOrderBST(level_order):
    print("No")
else:
    print("Yes")
```
